```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.lines as mlines
# 示例数据 (4组，每组2对，平均值和标准差)

plt.rcParams['font.size'] = 16 
means = [[-0.7702, -0.8749], [-0.7677, -0.5775], [-0.7129, -1.0069 ], [-0.9825, -0.9201]]
std_devs = [[0.4757, 0.2358], [0.347275, 0.3521], [0.2630, 0.2620], [0.3723, 0.2809]]
colors = ['blue', 'orange']
group_names = ['RBF', '$K_{MS}$', '$K_{\\alpha\\beta}$', '$K_{MS-AFS}$'] 
# 绘图
fig, ax = plt.subplots()

# 遍历每一组数据
for i, (mean, std) in enumerate(zip(means, std_devs)):
    x = [i * 2-0.3, i * 2 + 0.3]  # 每组的两个点
    for j in range(2):  # j 用于控制组内的两对数据
        ax.errorbar(x[j], mean[j], yerr=std[j], fmt='o', color=colors[j], capsize=5)
# 添加图例
# 创建虚拟的绘图对象用于图例

pair1_legend = mlines.Line2D([], [], color='blue', marker='o', linestyle='None', markersize=10, label='MTGP')
pair2_legend = mlines.Line2D([], [], color='orange', marker='o', linestyle='None', markersize=10, label='LMC')

# 自定义图例
ax.legend(handles=[pair1_legend, pair2_legend], loc='upper left', bbox_to_anchor=(1, 1))
ax.set_xticks([i * 2 for i in range(len(group_names))])  # 将每组的两个点中心对齐
ax.set_xticklabels(group_names)
# 美化图表
ax.set_xlabel('Kernel')
ax.set_ylabel('Negative Log-likelihood')
ax.set_title('Negative Log-likelihood on 2-shots subset')
plt.show()
plt.savefig('./paper/img/Negative Log-likelihood on 2-shots subset.png')
```
